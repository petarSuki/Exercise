```python
#Import all packages
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
#Import data
spam_df = pd.read_csv("/spam.csv")
```

```python
spam_df
```

```python
spam_df.groupby('Category').describe()
```

```python
#Turning spam/ham into numerical data and creating a new column called 'spam'
spam_df['spam'] = spam_df['Category'].apply(lambda x: 1 if x == 'spam' else 0 )
```

```python
spam_df
```

```python
# Creating train/test split
```

```python
x_train, x_test, y_train, y_test = train_test_split(spam_df.Message, spam_df.spam, test_size = 0.25)
```

```python
x_train.describe()
```

```python
# Finding word count and store data as matrix (storing data as numbers)
```

```python
cv = CountVectorizer()
x_train_count = cv.fit_transform(x_train.values)
```

```python
x_train_count.toarray()
```

```python
#Train model

model = MultinomialNB()
model.fit(x_train_count, y_train)
```

```python
# Pre-test normal mail
email_valid_mail = ["Hey wanna go for drinks?"]
email_valid_mail_count = cv.transform(email_valid_mail)
model.predict(email_valid_mail_count)
```

```python
# Pre-test spam
email_spam = ["reward money click free"]
email_spam_count = cv.transform(email_spam)
model.predict(email_spam_count)
```

```python
# Testing model

x_test_count = cv.transform(x_test)
model.score(x_test_count, y_test)
```
